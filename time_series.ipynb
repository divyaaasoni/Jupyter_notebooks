#### Time Series Components : 
*   Long term movement or direction present in the data
*   This represents underlying growth or decline in the series over time (extended period)
*   Trend can be upward, downward, stable/list.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

time_index=pd.date_range('2023-01-01','2023-12-31',freq='D')
trend_values=np.arange(len(time_index))
time_series_with_trend=pd.Series(trend_values,index=time_index)

plt.figure(figsize=(15,4))
plt.plot(time_series_with_trend)
plt.title('Time Series Data with Trend')
plt.xlabel('Time')
plt.ylabel('Values')
```

#### Seasonality
*   Regular / Predictable patterns which repeat over a specific period of time , often within a year or a day.
*   Seasonal effects are typically associated with calender related events, such as holiday seasons.
*   Retail sales might exhibit seasonality with increased activity during holiday seasons.

```python
seasonal_values=np.sin(2*np.pi*np.arange(len(time_index))/365)*10
time_series_with_seasonality=pd.Series(seasonal_values,index=time_index)

plt.figure(figsize=(15,4))
plt.plot(time_series_with_seasonality)
plt.title('Time Series Data with Seasonality')
plt.xlabel('Time')
plt.ylabel('Values')
plt.show()
```

#### Cyclic Pattern
*   Longer term pattern , and this not as strictly regular like seasonality.
*   These patterns may not have a fixed period and can span multiple seasons.
*   Cycles are often related to business cycle ,  daily sales , stock prices etc.

```python
cyclic_values=np.sin(2*np.pi*np.arange(len(time_index))/365)*15
time_series_with_cyclic=pd.Series(cyclic_values,index=time_index)

plt.figure(figsize=(15,4))
plt.plot(time_series_with_cyclic)
plt.title('Time Series Data with Cyclic')
plt.xlabel('Time')
plt.ylabel('Values')
plt.show()
```

#### Irregular Component


```python
irregular_values=np.random.normal(0,2,len(time_index))
time_series_with_irregular=pd.Series(irregular_values,index=time_index)

plt.figure(figsize=(15,4))
plt.plot(time_series_with_irregular)
plt.title('Time Series Data with Irregularity')
plt.xlabel('Time')
plt.ylabel('Values')
plt.show()
```
