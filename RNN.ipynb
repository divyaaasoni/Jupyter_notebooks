#### Recurrent Neural Network (RNN) :
-   Sequential Scenerios : where  input and output sequences are of same length. 
-   With RNN the feedback learning or the feedback loop works more efficiently . 
-   With each epoch running the model is constantly reciving feedbacks on the output layer. 

-   RNN is designed to handle sequential data more efficiently than ANN.
-   RNN processes input sequence of one element at a time.
-   RNN is more used into time series prediction.

```python
import numpy as np
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN , Dense
```

```python
np.random.seed(0)
xtrain=np.random.rand(100,10,1)
ytrain=np.random.rand(100,1)
xtest=np.random.rand(20,10,1)
ytest=np.random.rand(20,1)
```

```python
## Building the model
model=Sequential([
    SimpleRNN(64,input_shape=(10,1),activation='relu',return_sequences=False),Dense(1)
])
```

```python
model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)
```

-   MSE (Mean Squared Error) :  Squared up values of all the error , then finding the mean of all the squarred up errors. 
-   MAE (Mean Absolute Error) : The mean value of all errors in its absolute form . 

```python
model.fit(xtrain,ytrain,epochs=10,batch_size=32,verbose=1)
```

```python
test_loss , test_mae=model.evaluate(xtest,ytest,verbose=1)
print('Test MAE',test_mae)
```

```python
xnew=np.random.rand(5,10,1) # 5 new sequence
prediction=model.predict(xnew)
print('Prediction for new data:')
print(prediction)
```
